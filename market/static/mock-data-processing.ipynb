```python
rule_kind = r'(?P<Kind>.*) -'
rule_product = r'- (?P<Product>.*)'
```

```python
df['Kind'] = df['Name'].str.extract(rule_kind)
df['Product'] = df['Name'].str.extract(rule_product)
```

```python
df.head()
```

```python
df = df.drop(labels=['Name'], axis=1)
```

```python
df = df[["id", "Kind", "Product", "Price", "Description"]]
```

```python
df_dropped = df.dropna()
```

```python
df_dropped.head()
```

```python
price = r'\$(?P<Price>.*)'
df_dropped['New price'] = df_dropped['Price'].str.extract(price)['Price']
df_dropped['New price'] = df_dropped['New price'].astype(float).astype(int)
```

```python
df = df_dropped
```
